```python
import cv2
from matplotlib import pyplot as plt
```

```python
def plot_img(img):
    plt.imshow(img)
    plt.show()
```

```python
def img_cvt2gray(img):
    gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return gray_img
```

```python
def read_img(img):
    ip_img = cv2.imread(img)
    return ip_img
```

```python

imgR = read_img('assignment_5/right_img.jpg')# read right image
gray_r = img_cvt2gray(imgR)# convert to gray
imgL = read_img('assignment_5/left_img.jpg') # read left image
gray_l = img_cvt2gray(imgL)# convert to gray  

# initiste Stereo object 
stereo = cv2.StereoBM_create(numDisparities = 16, blockSize = 15)
 
# computes disparity
disparity_map = stereo.compute(gray_l, gray_r)
cv2.imwrite('assignment_5/disparity_img.jpg',disparity_map)
plot_img(disparity_map)
```
