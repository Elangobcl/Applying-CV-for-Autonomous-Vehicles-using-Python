```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("assignment_6/Student_Marksheet.csv")
print(df.keys())
```

```python
new_df = df.drop(['Student Name','Student Total'], axis=1)
new_df.head()
```

```python
new_df = new_df.drop(50)
```

```python
# Convert df to numpy
student_marksheet = new_df.to_numpy()
student_marksheet.shape
```

```python
mean_scores = np.mean(student_marksheet,axis=0) # calculate mean along column
```

```python
#Normalised Scores
normalised_scores = student_marksheet - mean_scores # calculate normalised score
normalised_scores
```

```python
#covariance Matrix
corr_scores = np.cov(normalised_scores.T)
corr_scores
```

```python
# calculate Eigen value and Eigen vector
Eigen_value, Eigen_vector = np.linalg.eig(corr_scores)
print ("Eigen_value : ", Eigen_value)
print ("Eigen_vector : ", Eigen_vector)
```

```python
# sort eigen value in descending order
descending_idx = np.argsort(Eigen_value)[::-1]
Eigen_value = Eigen_value[descending_idx]
Eigen_vector=Eigen_vector[descending_idx]
```

```python
projeced_matrix = Eigen_vector.T.dot(normalised_scores.T)
pca = projeced_matrix.T
pca=pca[:,:3]
pca
```
