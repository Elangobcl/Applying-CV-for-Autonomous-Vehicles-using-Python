```python
import cv2
from matplotlib import pyplot as plt
```

```python
def plot_img(img):
    plt.imshow(img)
    plt.show()
```

```python
def img_cvt2gray(img):
    gray_img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    return gray_img
```

```python
# read left and right images
imgR = cv2.imread('assignment-5/right.jpg')
gray_r = img_cvt2gray(imgR)
imgL = cv2.imread('assignment-5/left.jpg')
gray_l = img_cvt2gray(imgL)  
# initiste Stereo object 
stereo = cv2.StereoBM_create(numDisparities = 16, blockSize = 15)
 
# computes disparity
disparity = stereo.compute(gray_l, gray_r)
#plot_img(disparity)
cv2.imshow('output', disparity)
cv2.waitKey(0)
```
